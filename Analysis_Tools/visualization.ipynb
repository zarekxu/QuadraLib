```python
##----feature map visualization

import numpy as np
import cv2
import pickle as pl

import matplotlib.pyplot as plt

#---load feature maps from 1st order layer and second order layer, respectively
f1st = np.load("featuremap1st_1.npy")
f2nd = np.load("featuremap2nd_1.npy")
# print(f1st.shape)

a1 = np.mean(f1st[6], axis = 0)
a2 = np.mean(f2nd[6], axis = 0)
print(a1.shape)

b1=cv2.resize(a1,(32,32),interpolation=cv2.INTER_CUBIC)
b2=cv2.resize(a2,(32,32),interpolation=cv2.INTER_CUBIC)

plt.subplot(1, 2, 1)
plt.imshow((b1))
plt.colorbar()
plt.subplot(1, 2, 2)
plt.imshow((b2))
plt.colorbar()
```
